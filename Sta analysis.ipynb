```python
#import the data
df= pd.read_excel("social media data (1).xlsx")
```

```python
# Information about the data
df.info()
```

```python
# show the the number of Not a Number(NAN)
df.isna().sum()
```

```python
# Fill nan using forward fill method
df.fillna(method='ffill', inplace=True)

# Verify that the missing values have been filled
print(df.isna().sum())
```

```python
# what age column contains
df['Age '].unique()
```

```python
#Drop the Timestamp columns
df.drop(columns="Timestamp", inplace=True)
```

```python
# clean the age column data
df['Age '] = df['Age '].replace('21-25, 36-40', '21-25')

# Verify the updated 'Age' column
print(df['Age '].unique())
```

```python
# Assuming df is your DataFrame
df['Age '] = df['Age '].replace('21-25, 36-40', '21-25')

# Convert 'Age' to a categorical type with a custom ordering
age_order = ['15-20', '21-25', '26-30', '36-40']
df['Age '] = pd.Categorical(df['Age '], categories=age_order, ordered=True)

# Sort the DataFrame by the 'Age' column
df = df.sort_values('Age ')

# Verify the sorted 'Age' column
print(df['Age '].unique());
```

```python
# clean the Gender column
df['Gender '].unique()
```

```python
# Output of the updated data column
df['Gender '] = df['Gender '].apply(lambda x: 'Male' if 'Male' in x else 'Female')
# Verify the updated 'Gender' column
print(df['Gender '].unique())
```

```python
df["How often do you use social media "].unique()
```

```python
#Updated columns and renaming the column to Usage category
category_mapping = {
    'Multiple times a Day': 'Very Often',
    'Several times a week': 'Often',
    'Once a week': 'Rarely',
    'Once a Day': 'Rarely'}

# Applying mapping to the column
df['Usage Category'] = df['How often do you use social media '].map(category_mapping)
```

```python
#drop the 'How often do you use social media ' column
df.drop(columns=['How often do you use social media '], inplace=True)
```

```python
df.head()
```

```python
# Fill nan in the usage category using forward method
df["Usage Category"].fillna(method='ffill', inplace=True)
```

```python
#Updated usage category column
df["Usage Category"].unique()
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
social_media_columns = ['On average, how long do you use Facebook daily ?',
                        'On average, how long do you use Instagram daily ?',
                        'On average, how long do you use WhatsApp daily ?',
                        'On average, how long do you use LinkedIn daily ?']

# Plotting a bar chart
plt.figure(figsize=(10, 6))
for column in social_media_columns:
    plt.barh(column, df[column].mean(), label=column)

# Adding labels and legend
plt.xlabel('Social Media Platforms')
plt.ylabel('Average Daily Usage Time')
plt.title('Average Daily Usage Time on Social Media Platforms')
plt.legend()
plt.show()
```

### From our dataset it is obvious that the a very high percentage of our respondents make use of social media very often(86%). Also from the chart above, whatsapp is the social media platform with the highest usage, this could be accounted for based on the interface of the platform. Its features makes it easy to access, allows different modes of communications i.e voicenotes, images, videos, poll etc.When the need arises, lecturers can easily make use of whatsapp to explain concepts to students.

```python
df["Have you ever experienced negative feelings (stress, anxiety, depression) due to social media use?"].value_counts()
```

```python
df.groupby("Have you ever experienced negative feelings (stress, anxiety, depression) due to social media use?")["Usage Category"].value_counts().plot(kind="bar")
plt.xlabel("Impact of the usage")
plt.ylabel("Relative Frequency")
plt.title("Bar chart showing the impact of social media usage on mental health");
```

### Observing critically, we can infer from our data that the highest number of respondents fall in the group of those who use social media very often yet they are not affected negatively, this implies the students community may be well informed about preventive measures to take while using social media. The next highest group is the group of students who use the social media very often and are affected by their social media usage. This also communicates the idea that a very large percentage of the student's populace could not manage their use of social media platforms and are affected negatively.
